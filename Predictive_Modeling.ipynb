```python
# Importing necessary libraries
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Loading the Diabetes dataset
diabetes_data = load_diabetes(as_frame=True)
df = diabetes_data.frame
```

```python
# Displaying the first few rows of the dataset
df.head()
```

```python
# Creating a binary target variable by applying a threshold on the 'target' feature
# etting threshold to 140, meaning if value > 140, we classify as 1 (diabetes), else 0 (no diabetes)
df['diabetes_status'] = df['target'].apply(lambda x : 1 if x > 140 else 0)

# Droppimh the original target column 
df = df.drop(columns=['target'])
```

```python
df.head()
```

```python
# Splitting the data into features and target variable
x = df.drop(columns=['diabetes_status'])
y = df['diabetes_status']

# Splitting into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Initializing and training the RandomForestClassifier
model = RandomForestClassifier(random_state=42)
model.fit(x_train, y_train)
```

```python
# Making predictions on the test set
y_pred = model.predict(x_test)

# Calculating accuracy and generating classification report
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Displaying the results
print(f"Accuracy : {accuracy * 100:.2f}%")
print("\nClassification Report : \n", report)
print("\nConfusion Matrix : \n", conf_matrix)
```

```python
# Getting feature importance from the RandomForestClassifier
feature_importances = pd.DataFrame({'feature' : x.columns, 'importance' : model.feature_importances_})
feature_importances = feature_importances.sort_values(by='importance', ascending=False)

# Display feature importances
feature_importances
```
